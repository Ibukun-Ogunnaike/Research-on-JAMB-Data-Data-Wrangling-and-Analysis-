```python
#Grouping by year #average number of people that register for jamb rate by year
nth= data.groupby(["year"])["number"].aggregate("mean").reset_index()
print (nth)
```

From this we can observe that more people registerd for jamb in 2017 than 2018. We can't fully ascertain if this is the trend because of the limitation of our data.
